### JSON, DATATIME

json.load(...): json -> obj

json.dump(...): obj -> json


```python
import json
data = {'slogan': "Hello", "numbers": [1, 2, 3]}
with open("data.json", 'w', encoding= 'utf-8') as f:
    json.dump(data, f, ensure_ascii=False, indent=2)
```

```python
with open("data.json", 'r', encoding= 'utf-8') as f:
    data = json.load(f)
print(data)
```

DATATIME

```python
from datetime import datetime
dataToAdd = {'at': datetime.now().isoformat()}
with open("data.json", "w", encoding='utf-8') as f:
    json.dump(dataToAdd, f, ensure_ascii=False, indent=2)
    
with open("data.json", "r", encoding='utf-8') as f:
    obj = json.load(f)
    obj['at'] = datetime.fromisoformat(obj['at'])
print(obj["at"])
```

### CONTEXT MANAGER

Su dung context manager de dam bao thu hoi bo nho dung cach

```python
from contextlib import contextmanager
@contextmanager

def open_safely(path, mode='r', encoding='utf-8'):
    f = open(path, mode, encoding=encoding)
    try:
        yield f
    finally:
        f.close()

with open_safely('log.txt', 'w') as f:
    f.write('hello')
with open_safely('log.txt', 'r', )as f:
    text = f.read()
print(text)
```

### YIELD FROM

- Chuyen giao quyen kiem soat tu generator con ve generator cha

```python
def subGen():
    yield 1
    yield 2
def mainGen():
    result = yield from subGen()
    yield result
for val in mainGen():
    print(f"{val}")
```

- Ung dung vao list va dict

List

```python
def print(datas):
    yield "Start print..........."
    yield from datas
    yield "End print."

numbers = [1, 2, 3, 4, 5]
for item in print(numbers):
    print(item)
```

Dict

```python
def printDist(datass):
    yield "Start print..........."
    yield from datass.keys()
    yield from datass.values()
    yield "End point."
data = {"a": 1, "b": 2, "c": 3}
for item in printDist(data):
    print(item)
```

- combinations: Ket hop n phan tu voi nhau va loai bo di cac cap trung lap
- permuatations: Ket hop n phan tu voi nhau nhung khong loai bo di cac cap trung lap


- prodct: Gom tung phan tu thanh 1 nhom n phan tu ( co nhung phan tu va cac cap trung nhau)
- combinations_with_replacement: Gom cac phan tu thanh mot nhom n phan tu ( loai bo di cac cap trung nhau)


```python
char = ['a', 'b', 'c', 'd']
result1 = itertools.combinations(char, 2) # Ket hop 2 phan tu voi nhau (loai cac cap trung)
result2 = itertools.permutations(char, 2) # Ket hop 2 phan tu voi nhau (giu nguyen cac cap trung)
for re1 in result1:
    print(f"Combinations: {re1}")
    
for re2 in result2:
    print(f"Permutations: {re2}")
    
result3 = itertools.product(char, repeat=4)
for re3 in result3:
    print(f"Product: {re3}")
    
reuslt4 = itertools.combinations_with_replacement(char, 4)
for re4 in reuslt4:
    print(f"Combinations with replacement: {re4}")
```

- chain

```python
stt = [1, 2, 3, 4, 5]
name = ["Ti", "Toe", "Theo", "Tin", "Te"]
grade = [10, 10, 10, 10, 10]
combined = stt + name + grade
print(combined) # Result: [1, 2, 3, 4, 5, 'Ti', 'Toe', 'Theo', 'Tin', 'Te', 10, 10, 10, 10, 10]

#Co the su dung mot cach khac nhung van in ra ket qua nhu tren ----- Su dung itertools.chain()
combined2 = itertools.chain(stt, name, grade)
for item in combined2:
    print(item, end = ", ")
```

## Mo rong them du lieu

```python
import numpy as np
```

```python
d1 = {'Genre': ['Rock', 'Metal', 'Buluegraa', np.nan, np.nan, 'Rock'],
    'SocialMediaFollowers': [34345353, np.nan, 20000234000, 123342543, 17867876, np.nan],
      'SoldOut': [1, 0, 1, 0, 0, 1]}
d1
```

```python
df = pd.DataFrame(data = d1)
df
```

iloc (index location): dung de truy cap du lieu theo chi so hang va cot

```python
x = df.iloc[:, 0:2]
# ':' lay tat ca cac hang
# '0:2' lay tu cot 0-> 1
x
```

```python
y = df.iloc[:, 2] # Lay cot thu 3
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
numCols = ['SocialMediaFollowers'] # chua danh sach so luong nguoi theo doi
```

```python
catCols = ['Genre'] # chua danh sach cac cot du lieu dang phan loai
```

```python
from sklearn.pipeline import Pipeline
```

```python
from sklearn.preprocessing import StandardScaler
```

#### Cac buoc xu li la thay gia tri bi thieu bang trung binh cac gia tri, tiep theo la chuan hoa du lieu ve 1(std) && 0(mean)

```python
num_pipeline = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='mean')),
    ('scale', StandardScaler())
])
```

#### Tao categorical pipeline

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
cat_pipeline = Pipeline(steps= [
    ('impute', SimpleImputer(strategy='most_frequent')), # B1: impute
    ('one_hot_encoder', OneHotEncoder(handle_unknown='ignore', sparse_output=False)) # B2: chuyen category thanh cac cot nhi phan
])
cat_pipeline
```

```python
from sklearn.compose import ColumnTransformer
```

```python
col_transformer = ColumnTransformer(transformers= [
    ('num_pipeline', num_pipeline, numCols),
    ('cat_pipeline', cat_pipeline, catCols),
],
    remainder = 'drop',n_jobs = -1
)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()
```

```python
pipefinal = make_pipeline(col_transformer, dtc)
pipefinal
```

```python
pipefinal.fit(x_train, y_train)
```

```python
pipefinal.score(x_test, y_test)
```
